# Implement different Thresholding techniques on images using OpenCV

```python
import cv2
```

```python
img = cv2.imread("./inputs/zuck.png", cv2.IMREAD_GRAYSCALE)
```

```python
_, thresh_bin       = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)
_, thresh_bin_inv   = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV)
_, thresh_trunc     = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC)
_, thresh_toz       = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO)
_, thresh_toz_inv   = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV)
```

```python
cv2.imshow("Original", img)
cv2.imshow("Binary Threshold", thresh_bin)
cv2.imshow("Binary Threshold Inverted", thresh_bin_inv)
cv2.imshow("Truncated Threshold", thresh_trunc)
cv2.imshow("Threshold set to 0", thresh_toz)
cv2.imshow("Threshold set to 0 Inverted", thresh_toz_inv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
