# Write a program in python to Draw different geometric shapes and to write text on images using OpenCV.

```python
import cv2
import numpy as np
```

```python
h, w = 500, 500

img = np.zeros((h, w, 3), dtype=np.uint8)
```

```python
green = (0, 255, 0)
blue = (255, 0, 0)
red = (0, 0, 255)
yellow = (0, 255, 255)
```

```python
cv2.arrowedLine(img, (50, 50), (150, 150), green, 3)
```

```python
cv2.ellipse(
    img,
    center = (250, 250),
    axes=(100,100),
    angle = 0,
    startAngle = 0,
    endAngle = 360,
    color = blue,
    thickness = -1
)
```

```python
cv2.circle(img, center=(350,350),radius=50,color=red,thickness=3)
```

```python
x, y = 50, 250
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
cv2.putText(
    img,
    "Open",
    (x * 3, y),
    font,
    1.2,
    green,
    4,
    cv2.LINE_AA
)
cv2.putText(
    img,
    "CV",
    (x * 5, y),
    font,
    1.2,
    red,
    4,
    cv2.LINE_AA
)
```

```python
pts = (np.array([[40,400], [40,300], [100,350]], np.int32)).reshape((-1,1,2))
```

```python
cv2.polylines(img, [pts], True, yellow, thickness=2)
```

```python
cv2.imshow("CV Drawing", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
