# Write a program in python to perform various basic image processing operations: Reading image, writing image, conversion of images, and complement of an image using OpenCV.


```python
import cv2
```

```python
# Reading and Displaying the image

img = cv2.imread("./inputs/p1.webp")

cv2.imshow("image", img)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# Converting an image to GrayScale

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("gray image", img_gray)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# Finding complement of an image

img_comp = 255 - img

cv2.imshow("complement image", img_comp)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# Writing the image

cv2.imwrite("./outputs/p1_1.png", img)
cv2.imwrite("./outputs/p1_2.png", img_gray)
cv2.imwrite("./outputs/p1_3.png", img_comp)
```
