# Write a program in python to implement the various image filtering mechanisms using OpenCV

```python
import cv2
```

```python
img = cv2.imread("./inputs/zuck.png")
```

```python
mean_filtered = cv2.blur(img, (3, 3))
```

```python
median_filtered = cv2.medianBlur(img, 3)
```

```python
laplacian_filtered = cv2.convertScaleAbs(cv2.Laplacian(img, cv2.CV_64F))
```

```python
cv2.imshow("Original Colored Image", img)
cv2.imshow("Mean Filtered Image", mean_filtered)
cv2.imshow("Median Filtered Image", median_filtered)
cv2.imshow("Laplacian Filtered Image", laplacian_filtered)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
