# Write a program to demonstrate Haar feature-based cascade classifiers for Face and Eye Detection on images

```python
import cv2
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
img = cv2.imread("./inputs/zuck.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
for (x, y, w, h) in face_cascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30)):
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)

    for (ex, ey, ew, eh) in eye_cascade.detectMultiScale(gray[y:y + h, x:x + w]):
        cv2.rectangle(img[y:y + h, x:x + w], (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)
```

```python
cv2.imshow("Face and Eye Detection", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
