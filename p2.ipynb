# Write a program in python to resize, crop and rotate an image using OpenCV

```python
import cv2
```

```python
img = cv2.imread("./inputs/zuck.png")
```

```python
img_resized = cv2.resize(img, (200, 150))
```

```python
cv2.imshow("aisid", img_resized)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
h, w = img_resized.shape[:2]

img_cropped = img_resized[20:h-20, 40:w-40]
```

```python
h, w = img_cropped.shape[:2]


center = (w//2, h//2)

rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
img_rotated = cv2.warpAffine(img_cropped, rotation_matrix, (w, h))
```

```python
cv2.imwrite("./outputs/p2_1.png", img_resized)
cv2.imwrite("./outputs/p2_2.png", img_cropped)
cv2.imwrite("./outputs/p2_3.png", img_rotated)
```
