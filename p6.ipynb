# Write a program in python to perform different Morphological operations on images based on OpenCV

```python
import cv2
import numpy as np
```

```python
img = cv2.imread("./inputs/morph.jpg")

kernel = np.ones((5, 5), np.uint8)

img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1)
img_opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel, iterations=1)
img_closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel, iterations=1)
img_gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel, iterations=1)
```

```python
cv2.imshow("Original Image", img)
cv2.imshow("Eroded Image", img_erosion)
cv2.imshow("Dilated Image", img_dilation)
cv2.imshow("Opening Image", img_opening)
cv2.imshow("Closing Image", img_closing)
cv2.imshow("Gradient Image", img_gradient)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
