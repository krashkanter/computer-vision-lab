# Write a program in python to demonstrate image enhancement using contrast adjustment and histogram equalization using OpenCV

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("./inputs/zuck.png", cv2.IMREAD_GRAYSCALE)
```

```python
contrast_adjusted_image = cv2.convertScaleAbs(img, alpha=1.5, beta=0)
equalized_image = cv2.equalizeHist(img)
```

```python
cv2.imshow("Original Image", img)
cv2.imshow("Contrast Adjusted Image", contrast_adjusted_image)
cv2.imshow("Equalized Image", equalized_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Plotting

plt.subplot(2, 2, 1)
plt.hist(img.ravel(), 256, range=[0, 256])
plt.title("Original Image Histogram")

plt.subplot(2, 2, 2)
plt.hist(contrast_adjusted_image.ravel(), 256, range=[0, 256])
plt.title("Contrast Adjusted Image Histogram")

plt.subplot(2, 2, 3)
plt.hist(equalized_image.ravel(), 256, range=[0, 256])
plt.title("Equalized Image Histogram")

plt.tight_layout()
plt.show()
```
