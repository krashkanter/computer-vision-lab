# Write a program in python to demonstrate Bitwise Operations on Binary Images using OpenCV

```python
import cv2
```

```python
img1 = cv2.imread("./inputs/input1.jpg")
img2 = cv2.imread("./inputs/input2.jpg")
```

```python
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
```

```python
dest_and = cv2.bitwise_and(img1, img2)
dest_or = cv2.bitwise_or(img1, img2)
dest_xor = cv2.bitwise_xor(img1, img2)

dest_not1 = cv2.bitwise_not(img1)
dest_not2 = cv2.bitwise_not(img2)
```

```python
cv2.imshow("Image 1", img1)
cv2.imshow("Image 2", img2)

cv2.imshow("AND", dest_and)
cv2.imshow("OR", dest_or)
cv2.imshow("XOR", dest_or)
cv2.imshow("NOT Image 1", dest_not1)
cv2.imshow("NOT Image 2", dest_not2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
